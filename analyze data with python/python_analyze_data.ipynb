## this function get the average online buys for 5 weeks

```python
import array as arr

def resultAverage(result):
    print("average buy for 5 weeks:",result)
    
data = arr.array("i", [19,29,17,24,16])
dataList = data.tolist()

def getAverage(data):
    dataSum = sum(data)
    dataLen = len(data)
    average = dataSum / dataLen
    
    return average

averageResult = getAverage(dataList)

resultAverage(averageResult)
```

# Working with DataFrame

```python
# display the null value of the data column
marketingData[marketingData["Date"].isna()]
```

```python
marketingData = marketingData.dropna()
marketingData.isnull().sum()
```

```python
# display the summary statistics for all numeric column
marketingData.describe()
```

```python
# display the summary of a nominal data
marketingData["Day_Name"].describe()
```

```python
# select all the data when the day is monday & promo = "no promo"
marketingData[(marketingData["Day_Name"] == "Monday") & (marketingData["Promo"] == "No Promo")]
```

```python
# sort values by revenue in descending order so top 5 values are at top of the table
marketingData = marketingData.sort_values(by="Revenue", ascending=False)
marketingData.head()
```

## Aggregating the dataframe

```python
# sum of revenue by day name
# option 1 series
marketingData.groupby("Day_Name")["Revenue"].sum()
```

```python
# sum of revenue by day name
# option 2 dataframe
marketingData.groupby("Day_Name", as_index=False).agg({"Revenue": "sum"})
```

## Creating a new column from calculated fields

```python
# revenue per spend
marketingData["Revenue per Spend"] = marketingData["Revenue"] / marketingData["Marketing Spend"]
marketingData.head()
```

### 2. Collectiong the data

**The *_train_* dataset holds the sales data of an online office supplies and furniture shop from 2015 to 2018.**

### Clean the Data

```python
import pandas as pd

train = pd.read_csv("train.csv")
train.head()
```

```python
# basic methods for investigating/viewing a dataframe
train.info()
```

```python
train.isnull().sum()
```

```python
train[train["Postal Code"].isna()]
```

```python
# filling in null values with non-null values
train["Postal Code"] = train["Postal Code"].fillna("095401")
train.info()
```

```python
train.isnull().sum()
```

```python
# deleting a column
del train["Row ID"]

train.head()
```

```python
# checking/converting data types
train.info()
```

```python
# converting object to datetime
train["Order Date"] = pd.to_datetime(train["Order Date"], format="%d/%m/%Y")
train["Ship Date"] = pd.to_datetime(train["Ship Date"], format="%d/%m/%Y")

# converting float to object/string
train["Postal Code"] = train["Postal Code"].astype(str)

train.info()
```
