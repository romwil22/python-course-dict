## this function get the average online buys for 5 weeks

```python
import array as arr

def resultAverage(result):
    print("average buy for 5 weeks:",result)
    
data = arr.array("i", [19,29,17,24,16])
dataList = data.tolist()

def getAverage(data):
    dataSum = sum(data)
    dataLen = len(data)
    average = dataSum / dataLen
    
    return average

averageResult = getAverage(dataList)

resultAverage(averageResult)
```

# Modeling data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## _What is the yearly sales trend?

```python
train = pd.read_csv("train.csv")

train.info()
train.head()
```

```python
# converting object to datetime
train["Order Date"] = pd.to_datetime(train["Order Date"], format="%d/%m/%Y")

# create order year column
train["Order Year"] = train["Order Date"].dt.year
train.info()
train.head(2)
```

```python
train["Order Year"] = train["Order Year"].astype(str)

train.info()
```

```python
# sum of sales by year
yearSale = train.groupby("Order Year", as_index=False).agg({"Sales": "sum"})

yearSale
```

```python
#graphing data
sns.lineplot(x="Order Year", y="Sales", data=yearSale, marker="o")
plt.xlabel("")
plt.show()
```

## Q.1 Answer: the yearly sales seems to be showing an upward trend.

## _Q2. Which product sold the most in 2018_

```python
sales2018 = train[train["Order Year"] == "2018"]

sales2018.head()
```

```python
product = sales2018.groupby("Product Name", as_index=False).agg({"Sales": "sum"})
prod = sales2018.sort_values(by="Sales", ascending=False)

prod.head()
```

## Q.2 Answer: The top selling product in 2018 is Canon imageCLASS 2200 Advanced Copier

```python
prodTop = prod[prod["Sales"] > 4416.174]

# graphing data
sns.barplot(y="Product Name", x="Sales", data=prodTop, ci=False, orient="h", color="blue")
plt.title("Top product in 2018")
plt.ylabel("")
plt.show()
```

```python
sns.barplot(y="Product Name", x="Sales", data=prodTop, orient="h", color="blue")
plt.title("Top product in 2018")
plt.ylabel("")
plt.show()
```

## _Q3. Among the three product categories, which one had the lowest sales in 2018?_

```python
productSale = sales2018.groupby("Category", as_index=False).agg({"Sales": "sum"})

productSale
```

## Q.3 Answer: Among furniture's Sales scored the lowest in 2018.

```python
# graphing data

sns.barplot(x="Category", y="Sales", data=productSale, palette="Blues")
plt.title("Sales by product category")
plt.xlabel("")
plt.ylabel("")
plt.show()
```

## 4. What's the yearly sales trend of the three customer groups?

```python
consumerSegment = train[train["Segment"] == "Consumer"]
consumer = consumerSegment.groupby("Order Year", as_index=False).agg({"Sales": "sum"})

consumer
```

```python
homeSegment = train[train["Segment"] == "Home Office"]
home = homeSegment.groupby("Order Year", as_index=False).agg({"Sales": "sum"})

home
```

```python
corporateSegment = train[train["Segment"] == "Corporate"]
corporate = corporateSegment.groupby("Order Year", as_index=False).agg({"Sales": "sum"})

corporate
```

```python
# graphing data
sns.lineplot(x="Order Year", y="Sales", data=consumer, marker="o")
sns.lineplot(x="Order Year", y="Sales", data=home, marker="o")
sns.lineplot(x="Order Year", y="Sales", data=corporate, marker="o")

plt.legend(["consumer", "home", "corporate"], fontsize=8)
plt.show()
```

### Q.4 Answer: All of the three customer  groups exhibited an upward trend starting in 2016, but the corporate customers had quite a big in 2017.

#### Q.5 Answer: We can say that the number of corporate customers is increasing yearly.
