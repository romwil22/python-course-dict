## this function get the average online buys for 5 weeks

```python
import array as arr

def resultAverage(result):
    print("average buy for 5 weeks:",result)
    
data = arr.array("i", [19,29,17,24,16])
dataList = data.tolist()

def getAverage(data):
    dataSum = sum(data)
    dataLen = len(data)
    average = dataSum / dataLen
    
    return average

averageResult = getAverage(dataList)

resultAverage(averageResult)
```

## pandas series, one dimensional object

```python
import pandas as pd
```

```python
dItems = pd.Series(["chairs","labels","tables"])
dItems
```

```python
# creating series with a in list with the index
dItems = pd.Series(["chairs","labels","tables"], index=["ORDER 1","ORDER 2","ORDER 3"])
dItems
```

```python
# creating a sereis by passing in a list without the index
dName = pd.Series(["rom","mheann","ana"])
dName
```
